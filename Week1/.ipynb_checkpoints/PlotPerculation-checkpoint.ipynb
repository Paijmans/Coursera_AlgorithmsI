# Plot perculation threshold geerated using C++ QuickUnion Code.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
percs = np.genfromtxt('perculation_100.dat', delimiter=',')
sample_nbr = len(percs)
number, boundaries = np.histogram(percs, bins=1000, range=(0.,1.))
cum_perculation_frac = number.cumsum()/sample_nbr
```

## Perculation with 20x20 grid.

```python
print("Perculation threshold: " + str(np.mean(percs)) + " +/- (" + str(np.std(percs)) + ")")

plt.vlines(np.mean(percs), 0, 1, colors='black', linestyles=':')
plt.plot(boundaries[1:], cum_perculation_frac)
```

## Perculation with 100x100 grid.

```python
print("Perculation threshold: " + str(np.mean(percs)) + " +/- (" + str(np.std(percs)) + ")")

plt.vlines(np.mean(percs), 0, 1, colors='black', linestyles=':')
plt.plot(boundaries[1:], cum_perculation_frac)
```
